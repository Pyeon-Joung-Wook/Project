# 연도별 시트 합한 결과

```python
file_path = "C:/Users/elris/Python/Untitled Folder/코스피200 편입여부.xlsx"
df_list = []
for year in range(2010,2020) : 
    kospi_df = pd.read_excel(file_path, sheet_name= str(year)+"-01-31")   # 2011-01-31, 2012-01-31, ...
    code_list = []
    for code in kospi_df['Symbol']: 
        code_list.append(str(code).replace('A',''))
    kospi_df['Symbol'] = code_list
    df_list.append(kospi_df)
```

```python
for i in df_list : 
    display(i.head())
```

```python
for num, code in enumerate(df_list) : 
    if num == 0 : 
        group = set(code['Symbol'])
    else : 
        group = set(set(group).union(set(code['Symbol'])))
```
