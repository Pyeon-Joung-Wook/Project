```python
plt.figure(figsize = (20,20))

color_list = ['#800000', '#FFFF00', '#808000', '#008080', '#000080', '#800080',  '#FF69B4', '#4B0082', '#B0C4DE', '#48D1CC', '#BC8F8F','#FF6347']
i = 0
for c in set(crowding_list['종목 코드']):
    temp = crowding_list[crowding_list['종목 코드']==c]
    plt.scatter(x = temp['달성율'],
               y = temp['서포터 수'],
               s = temp['총액']/100000,
               c = color_list[i],
               label= set(temp['종목']),
               alpha=0.5, edgecolors='none')
    i= i+1

plt.title("종목 별 서포터 수 및 달성율의 관계", fontsize = 30)
plt.xlabel('달성율', fontsize = 20)
plt.ylabel('서포터 수', fontsize = 20)

plt.legend()
plt.grid()
plt.show()
```
